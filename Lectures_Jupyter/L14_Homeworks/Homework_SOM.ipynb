```python
import numpy as np
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
```

In this exercise we are going to place 100 points in the map:

```python
h,w, maxit=500,500,50
"""Returns an image of the Mandelbrot fractal of size (h,w)."""
X = np.ones((500,500))
c = (-0.8+0.156*1j)*X
y,x = np.ogrid[ -1.5:1.5:h*1j, -1:1:w*1j ]
z= x+y*1j


Julia = maxit + np.zeros(z.shape, dtype=int)

for i in tqdm_notebook(range(maxit)):
    z = z**2+c 
    diverge = (z)*np.conj(z) > 2**2            # who is diverging
    div_now = diverge & (Julia==maxit)  # who is diverging now
    Julia[div_now] = i                  # note when
    z[diverge] = 2                        # avoid diverging too much


fig = plt.figure()
ax = fig.add_axes([0., 0., 1., 1., ])
# Hide grid lines
ax.grid(False)
ax.axis('off')
# Hide axes ticks
plt.imshow(Julia)


plt.show()
```

And they will cluster around the more colorful points using SOM.

```python
points = []
for x_values in range(0,500,50):
    for y_values in range(0,500,50):
        points.append([[x_values,y_values],Julia[x_values,y_values]])
```

```python
MatrixOfDifferences = [[abs(point[1]-otherPoint[1]) for point in points] for otherPoint in points]
```
